The most transformative companies often emerge from fundamental research that seemed commercially irrelevant at the time. Google came from a PhD project. Genentech came from recombinant DNA research. mRNA vaccines came from decades of basic science. But traditional venture capital isn't structured for this kind of long-horizon investment. This post explores the economics of deep tech and patient capital.



## What Is "Deep Tech"?

The term "deep tech" lacks precise definition, but the pattern is recognizable:

**Technical risk dominates market risk**: Traditional startups face market risk—will customers buy? Deep tech startups face technical risk—can we make this work at all?

**Long development timelines**: Years or decades from research to product. Drug development takes 10-15 years. Fusion power has been "20 years away" for 60 years.

**Capital intensity**: Hardware, labs, clinical trials, and manufacturing facilities cost far more than software development.

**Intellectual property matters**: Patents, trade secrets, and know-how create barriers that might last longer than network effects.

Examples: autonomous vehicles, quantum computing, nuclear fusion, advanced materials, biotech, robotics, space technology.



## Historical Examples

**Genentech**: Founded 1976 to commercialize recombinant DNA technology (gene splicing). First product (synthetic insulin) approved 1982. IPO 1980. Acquired by Roche in 2009 for $46B. From research to massive outcome: ~30 years.

**Google**: Started as a Stanford PhD project in 1996. Took years to find a business model (AdWords launched 2000). Became the most valuable company by market cap. From research to dominance: ~10 years.

**Tesla**: Founded 2003. First profitable quarter: Q3 2019. Became one of the most valuable car companies. Path from startup to profitability: ~16 years of losses and skepticism.

**SpaceX**: Founded 2002. First successful orbital launch: 2008. First Falcon 9 reuse: 2017. First Starship test: 2023. Path to operational rocket: ~15 years.

These are survivorship-biased examples. For every Tesla, many deep tech companies failed or limped along. But the successes are enormous.



## The Funding Gap

Deep tech startups often hit a "valley of death" between research and commercialization:

**Grants and seed**: Universities and early-stage investors fund initial research. Amounts: $100K-$1M.

**The gap**: Technology is promising but unproven. Too early for traditional VCs who want revenue traction. Too speculative for corporate partners. Need ~$5-50M to get to next milestone.

**Series B and beyond**: If the technology works, traditional VCs might join. But getting there is the hard part.

This gap is particularly acute for capital-intensive hardware. Software startups can reach product-market fit with $2-5M. Hardware startups might need $50M just for a prototype.



## Why VC Isn't Structured for This

Traditional VC funds have structural mismatches with deep tech:

**10-year fund life**: VC funds typically have 10-year terms. Deep tech might need 15-20 years. This creates pressure to exit earlier than optimal.

**Power law focus**: VCs want companies that can scale quickly to massive outcomes. Deep tech often has longer, more linear growth curves.

**Portfolio model**: VCs expect most investments to fail and aim for 1-2 massive wins. Deep tech failure modes are slower—you may not know for 5 years if it's working.

**Partner expertise**: Most VCs are trained in software and business models, not physics, biology, or hardware. Evaluating deep tech requires different expertise.

**Follow-on dynamics**: Deep tech needs multiple large rounds over many years. VCs need to reserve capital for follow-ons, limiting how much they can deploy initially.



## Alternative Structures

Several models attempt to bridge the gap:

**ARPA-E and government programs**: US Department of Energy's Advanced Research Projects Agency-Energy funds high-risk energy research. DARPA does similar for defense. Grants without equity, long timelines.

**Corporate labs**: Bell Labs, Xerox PARC, DeepMind (before Google acquisition). Corporations can have longer time horizons than VCs. But corporate labs face their own pressures (quarterly earnings, strategic shifts).

**Evergreen funds**: Some firms (Flagship Pioneering, Lux Capital, Breakthrough Energy Ventures) raise capital with longer horizons and are structured for patient investment.

**Family offices and sovereign wealth**: Patient capital from sources not subject to 10-year fund cycles. Can invest with 20-30 year horizons.

**Project-specific financing**: Some deep tech (mining, energy, manufacturing) can raise project finance against future revenue, not equity.



## Sector-Specific Timelines

Different deep tech sectors have different dynamics:

**Biotech**: Well-established venture model exists. Drug development is long (10-15 years) but milestones are clear (phases 1-2-3 approval). IPO window allows early investor exits even before product launch.

**Robotics**: Hardware is hard; software is differentiating. Path to market might be 5-10 years. Manufacturing scale is challenging.

**Advanced materials**: Very long development cycles. Batteries, solar, carbon capture. Often need manufacturing at scale to prove economics.

**Quantum computing**: Still in "science experiment" phase for many applications. Useful quantum advantage might be 10-20 years away for general use.

**Fusion**: Has been "20 years away" for decades. Recent private investment suggests some believe breakthrough is closer. But timelines are inherently uncertain.



## Valuation Challenges

How do you value a company with:
- No revenue
- No product
- Technical risk unresolved
- 10-year path to commercialization?

Traditional methods fail:
- **DCF**: How do you discount cash flows that might start in 15 years with high probability of zero?
- **Comparables**: There may be no comparable companies.
- **Revenue multiples**: No revenue to multiply.

In practice, deep tech valuation is:
- **Option pricing**: Value the technology as a call option on future markets.
- **Milestone-based**: Value increases as technical risks are retired.
- **Comparable transactions**: What did other acquirers pay for similar technology?

This is closer to R&D portfolio management than traditional startup investing.



## The Talent Market

Deep tech is constrained by talent as much as capital:

**PhDs**: Many deep tech companies need PhDs in specialized fields. These people are rare. Universities produce limited numbers. Big tech competes for them.

**Location matters**: Talent clusters around universities and labs. Boston (biotech), SF (AI), Austin (semiconductors). Deep tech companies often can't be fully remote.

**Immigration**: Many deep tech founders and employees are immigrants. Visa restrictions limit talent supply (in the US).

**Compensation expectations**: Top technical talent can command $500K+ in big tech. Startups compete with equity, but early deep tech equity is highly uncertain.



## Portfolio Construction for Deep Tech

Given the dynamics, how should investors construct deep tech portfolios?

**More shots on goal**: Given high technical risk, diversification matters. But each shot is expensive. Trade-off between diversification and depth of investment.

**Stage specialization**: Some investors focus on early stage (pre-technical proof); others on later stages (technology works, now commercialize). Different skills and capital structures required.

**Sector expertise**: Deep tech investment requires domain expertise to evaluate technical risk. Generalist VCs struggle.

**Long reserve ratios**: Deep tech needs follow-on capital. Investors must reserve more for follow-ons than in software.

**Portfolio patience**: Expect longer holds. Returns may take 15+ years to materialize.



## Why It Matters

Deep tech matters because:

**Transformative outcomes**: The biggest changes to how we live come from fundamental technology—computing, biotech, energy. Software builds on these foundations.

**Long-term competitiveness**: Countries that invest in deep tech research maintain industrial leadership. China, the US, and Europe are competing for this position.

**Solutions to hard problems**: Climate change, disease, food security—these require deep tech, not just better apps.

**Career opportunity**: For researchers and engineers, deep tech offers the chance to work on hard, meaningful problems—not just optimizing ad clicks.

The mismatch between the importance of deep tech and the structure of capital to fund it is one of the significant market failures in innovation finance. Fixing it would accelerate progress on problems that matter.
