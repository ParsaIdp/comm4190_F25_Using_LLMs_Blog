Raising venture capital is opaque. Founders get rejections without understanding why. VCs seem to pattern-match in mysterious ways. But there's more structure than it appears. Understanding how VCs actually evaluate companies—especially AI companies—helps founders prepare, pitch, and navigate the process.



## The VC Business Model

To understand VC decisions, understand the incentives.

**Power laws**: VC returns are dominated by a tiny fraction of investments. The best deal in a fund might return more than all other deals combined. VCs optimize for finding outliers, not avoiding failures.

**Portfolio construction**: A typical fund makes 20-40 investments. Most will fail or return 1-2×. The fund needs 1-3 massive wins (10-100×) to hit its return target.

**Fund economics**: VCs charge ~2% annual management fee and ~20% of profits (carry). Limited partners (LPs) expect returns of 3× their capital over 10 years. Anything less, and LPs won't re-invest.

**Implications for founders**:
- VCs won't invest in "good companies" that can't be huge. You need a path to 10× or more.
- VCs would rather miss a good company than invest in a mediocre one. Passing is easy; regret is painful but infrequent.
- VCs are looking for conviction about exceptional outcomes, not certainty about moderate ones.



## Evaluation Heuristics

VCs claim to evaluate many things. In practice, a few dimensions dominate:

**Team**: Who are you? Do you have relevant experience? Do you seem capable of executing? Are you coachable? Do you have technical depth in AI if it's an AI company?

For AI companies: VCs want to see that founders understand the technology, not just the application. Can you speak fluently about models, training, inference, and the frontier? If you can't, you need a technical co-founder who can.

**Market**: Is this market large enough to produce a huge company? Is it growing? Is it winnable by a startup or controlled by incumbents?

For AI companies: Markets that didn't exist before AI (autonomous vehicles, protein design) are attractive. Markets where AI is a feature (adding AI to existing software categories) are harder—incumbents can add features.

**Traction**: What have you built? Who's using it? Are they paying? Are they retaining? How fast is it growing?

For AI companies: Demos don't count. VCs have seen too many impressive demos that collapse in production. Show real usage, preferably with revenue.

**Differentiation**: What do you have that others don't? Why will you win against competitors (including future ones)?

For AI companies: This is the crux. "We have a better model" is weak—models commoditize. "We have proprietary data" or "We own the workflow" is strong.



## AI-Specific Evaluation

AI companies get additional scrutiny:

**Model ownership**: Do you train your own models or use APIs? Own models = more differentiation but more cost. API-dependent = lower cost but thin differentiation.

**Data strategy**: Where does your training/fine-tuning data come from? Is it proprietary? Does the product generate more? Data is the most discussed form of moat in AI.

**Technical depth**: Can the team actually innovate at the model level, or are they "prompt engineers"? The deepest expertise commands the most respect.

**Moat durability**: VCs are paranoid about OpenAI/Google/Meta releasing capabilities that destroy your differentiation. You need a story for why you'll stay ahead or why the race doesn't matter (because you own something they don't).

**Unit economics in AI**: Compute costs are high. VCs want to see that you can deliver value profitably. If you spend $1 on inference for every $0.50 in revenue, you're in trouble.



## The Partner Meeting

Inside a VC firm, decisions are made in partner meetings. Here's how they typically work:

**Champion**: One partner (the one who met you) presents the deal to others. They've already done initial diligence and are advocating for investment.

**Discussion**: Other partners ask questions, poke holes, share concerns. Standard objections: "market too small," "why them," "seen this before," "too early/late."

**Vote or consensus**: Different firms have different decision processes. Some require unanimous partner approval; others allow individual partner conviction.

**Time pressure**: Competitive deals (multiple VCs interested) force faster decisions. Slow deals with no competition allow more diligence—and often lead to passes.

For founders:
- Your champion needs ammunition to defend you. Give them clear, memorable points to make in the meeting.
- Anticipate the objections. "Why can't OpenAI do this?" has to be answered.
- The partner meeting is the decision point. Everything before is filtering.



## Pattern Matching

VCs pattern-match, consciously or not. They recognize archetypes:

**Positive archetypes**:
- "Second-time founder who learned from previous attempt"
- "Deep technical expert from top lab who sees opportunity"
- "Industry insider who understands the problem better than anyone"
- "Former big-tech PM who knows how to scale products"

**Negative archetypes**:
- "First-time founder who's never shipped anything"
- "Business person pitching technical product without technical co-founder"
- "Pivot from unrelated domain with no evident expertise"
- "Big vision, no traction, been at it for two years"

This pattern-matching is both rational (based on evidence about what predicts success) and biased (based on limited samples and flawed heuristics). It's imperfect, but it's real.



## Red Flags and Green Flags

**Red flags** that kill deals:
- Founders who can't articulate why they'll win
- No answer to "what's your differentiation?"
- Defensiveness about technology or traction questions
- Unrealistic financial projections with no grounding
- Co-founder conflict or recent departures
- Dependent on single customer or single model provider

**Green flags** that advance deals:
- Founders who demonstrate clear thinking about market dynamics
- Strong retention or engagement metrics (even at small scale)
- Unusual insight about customers or technology
- Existing relationships that will become customers
- Team with previous working history together
- Competitive process (other VCs are interested)



## Due Diligence

If a partner meeting goes well, due diligence follows:

**Technical diligence**: VCs (or their advisors) evaluate your technology. Can you actually do what you claim? Is the architecture sound? Is the team capable?

For AI companies: Be ready to walk through model architecture, training methodology, evaluation results, and roadmap. "It's a transformer" isn't enough. What did you do that's hard?

**Commercial diligence**: Talk to customers. Are they actually using the product? What do they say about you? Would they pay more?

**Reference checks**: The VC calls people who've worked with you. This is where skeletons emerge. Be honest about your history.

**Legal diligence**: Cap table, IP ownership, employment contracts, data rights. Make sure your house is in order before it's inspected.



## What Founders Control and Don't

**You control**:
- Clarity of your pitch
- Quality of materials (deck, demo, data room)
- Your responsiveness and professionalism
- How you handle objections and follow-ups
- Preparation and knowledge of your own business

**You don't control**:
- Whether your market matches their thesis
- Their fund dynamics (too early in fund, too late, already invested in competitor)
- Partner dynamics (who champions you, who's skeptical)
- Macro conditions (fundraising environment, public market valuation comparables)

Many rejections have nothing to do with you. Funds pass on good companies for strategic reasons. Don't take every "no" as a judgment on your startup.



## Advice for Founders

1. **Run a tight process**: Talk to many VCs in parallel. Create competitive pressure. Don't go exclusive early.

2. **Know your numbers**: Metrics, unit economics, growth rates. VCs respect founders who know their business cold.

3. **Tell a story**: Not just what you're building, but why now, why you, and what's the end state. Narratives stick.

4. **Anticipate the AI-specific questions**: What's your moat? What if OpenAI ships this? Why can't you be replaced by a prompt?

5. **Get introductions**: Cold outreach rarely works. Warm introductions from trusted sources are the best path to partner meetings.

6. **Know when to walk**: Some VCs will waste your time. If they're slow, unresponsive, or asking for unreasonable diligence without commitment, move on.

Fundraising is a sales process. Treat it systematically, not emotionally.
