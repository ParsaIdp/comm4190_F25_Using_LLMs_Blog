The AI startup boom has produced thousands of companies across every layer of the stack. Some will become giants. Most will die or be acqui-hired. The difference often isn't technical capability—it's strategic positioning. Where in the value chain are you? What makes your position defensible? Can you build a business that persists once the technology is commoditized?



## The AI Value Chain

Think of AI as a stack with value distributed across layers:

**Infrastructure**: Compute, chips, cloud, training infrastructure. Players: NVIDIA, cloud providers, CoreWeave. High capex, high margins for winners, brutal for losers.

**Models**: Foundation models, fine-tuning, specialized models. Players: OpenAI, Anthropic, Google, Meta (Llama). High R&D costs, network effects from developer ecosystems.

**Tools and platforms**: APIs, orchestration, evaluation, deployment. Players: LangChain, Weights & Biases, Hugging Face. Developer adoption as the moat.

**Applications**: End-user products built on AI. Players: thousands of startups, plus incumbents adding AI features. Closest to customers, but often dependent on layers below.

Value concentration shifts over time. Early, models captured most value. As models commoditize, applications and infrastructure may capture more. Figuring out where value will concentrate is the key strategic question.



## Commoditization Pressure

The base models are commoditizing faster than expected:

**Open-source models**: LLaMA, Mistral, and others have reached near-parity with proprietary models for many tasks. If the best model is free, the API provider can't charge a premium.

**API price wars**: OpenAI, Anthropic, and Google are cutting prices rapidly. Input tokens went from $0.03 to $0.001 in two years. Inference is becoming cheap.

**Low switching costs**: Standardized prompting interfaces make it easy to switch between providers. If Claude and GPT-4 are interchangeable for your application, neither can charge a premium.

This is great for builders but terrifying for model providers. Commoditization pressure flows up the stack: if models are cheap, model providers need volume, which requires developer tools, which requires applications.



## Defensibility Patterns

What makes an AI startup defensible once the technology is no longer differentiating?

**Data moats**: Proprietary data that's hard to replicate. Bloomberg has financial data; OSHA has regulatory data; hospitals have medical records. If your product generates unique, valuable data, that's a moat.

**Workflow embedding**: Products that become part of daily workflow are sticky. Once your tool is integrated into CI/CD pipelines, document workflows, or customer service systems, switching is expensive.

**Vertical integration**: Own the whole experience in a niche. Vertical AI companies can optimize across the stack for their specific domain—better than horizontal providers can.

**Network effects**: Two-sided marketplaces, collaboration tools, or social products where value increases with users. Hard to create in AI but powerful when achieved.

**Speed and execution**: In early markets, being first with a good-enough product beats being late with a perfect product. But this is temporary; execution alone isn't a moat.

**Regulatory capture**: In regulated industries, existing players have relationships, approvals, and compliance infrastructure. AI startups can inherit this by partnering or acquiring.



## Failure Modes

**"Thin wrapper" syndrome**: Your product is an API call with a UI. You've added prompts, maybe a workflow, but no real value. Model providers can replicate your product trivially.

When OpenAI ships plugins and GPTs, when Anthropic improves Claude, your "thin wrapper" disappears into their platform. You didn't build defensibility.

**Model dependency**: Your product depends on one model provider. They raise prices, change terms, or deprecate the model you're using. You're exposed to their roadmap.

**CAC/LTV inversion**: Customer acquisition cost exceeds lifetime value. Easy to do in AI: pay for marketing, acquire users, but users churn when the novelty wears off.

**Demo-to-production gap**: Your demo is impressive, but reliability in production is terrible. Hallucination rates, latency, cost, and edge cases kill you in real deployment.

**Commoditization from above**: You build a specialized AI tool; the foundation model provider absorbs your functionality into their base offering. Game over.



## Vertical vs. Horizontal

**Horizontal plays** sell to everyone: AI writing assistants, code completion, general chatbots. The market is huge, but competition is fierce and differentiation is hard.

**Vertical plays** focus on specific industries: AI for law firms, AI for healthcare billing, AI for insurance claims. The market is smaller, but:
- Less competition
- Domain expertise creates barriers
- Customers pay more for specialized solutions
- Regulatory and compliance requirements lock competitors out

The best vertical plays combine AI with deep domain understanding. You're not "AI for X"—you're "the best solution for X, powered by AI." The AI is an implementation detail, not the value proposition.



## Timing

**Too early**: You're building on technology that isn't ready. The AI can't reliably do what you need. You burn capital waiting for capability to catch up.

Being too early is worse than being too late. At least late entrants know what works.

**Too late**: Incumbents have added AI features. Startups have established market position. You're fighting for a slice of a pie that's already divided.

**Just right**: Technology is capable, market is forming, incumbents are slow to adapt, but you're not so early that you're the only buyer. This window is narrow and hard to identify in real-time.

Timing luck is underrated in startup success. Many failures and successes differ less in strategy than in when they entered.



## Case Studies: Winners and Losers

**Jasper AI**: Early mover in AI copywriting. Grew fast but faced commoditization as ChatGPT ate their use case. Has pivoted towards enterprise marketing workflows. Verdict: still fighting, outcome uncertain.

**Figma AI features**: Incumbent adding AI to existing product. Figma has the workflow, the users, and the data. AI is a feature, not a separate product. Verdict: likely durable.

**Harvey AI**: AI for legal work. Deep vertical integration, domain expertise, relationships with major firms. Building specific models on proprietary legal data. Verdict: defensible position.

**Notion AI**: AI added to existing productivity tool. Notion has the workflow; AI enhances it. Users won't switch to "Notion + separate AI tool." Verdict: durable for existing users.

Pattern: Startups that owned workflow first, then added AI, tend to fare better than startups that built "AI wrapper" products hoping to capture workflow later.



## What VCs Actually Look For

Talking to VCs reveals consistent patterns:

**Team**: Deep domain expertise or strong technical credentials. Founders who can articulate why they specifically will win.

**Market**: Large enough to matter ($1B+), with clear path to capture meaningful share. Not "AI for everything"—focused opportunity.

**Traction**: Real usage, preferably with paying customers. In AI, demos are easy; production deployment is hard. Show the hard part.

**Differentiation**: What makes you not a thin wrapper? Data, workflow, technology, go-to-market, partnerships?

**Unit economics**: Can you acquire customers profitably? What's CAC and LTV? In early stage, these may be projections, but the logic should hold.

**OpenAI-proofing**: VCs ask: "What happens when Sam ships this feature?" You need an answer. The best answer is that you're in a vertical or have an asset OpenAI won't build.



## My Playbook for AI Startups

If I were starting an AI company today:

1. **Go vertical**: Pick a specific industry where I have expertise or can acquire it quickly.

2. **Own the workflow**: Build a product that structures how users do their job, with AI enhancing that workflow. Not "AI tool" but "better way to do X."

3. **Generate proprietary data**: Design the product so that usage creates valuable data that improves the product. The flywheel is the moat.

4. **Model-agnostic when possible**: Don't depend on one provider. Abstract the AI layer so you can switch.

5. **Charge for value, not API calls**: Price based on outcomes or seats, not tokens. Don't let customers see the underlying commodity.

6. **Move fast**: The window for capturing position is short. Speed beats perfection.

None of this is novel advice. The hard part is execution.
